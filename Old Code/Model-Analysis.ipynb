```python
# Importing libraries and dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 
import warnings
from sklearn.impute import SimpleImputer # used for handling missing data (numerical)
from sklearn.preprocessing import LabelEncoder, OneHotEncoder # used for encoding categorical data
from sklearn.model_selection import train_test_split # used for splitting training and testing data
from sklearn.preprocessing import StandardScaler # used for feature scaling
warnings.filterwarnings('ignore')

# Assign data to df const
df = pd.read_csv('../Datasets/Combined_Dataset.csv')
new_state_list = []
for state in df['State']:
    new_state_list.append(state)
print(new_state_list)
```

```python
# Splitting the attributes into independent and dependent attributes
X = df.iloc[:, :-1].values # attributes to determine depndent variable / Class
Y = df.iloc[:, -1].values # dependent variable / Class
```

```python
# Handling missing numerical or cateogrical data
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent') # Replacing NaN numerical data with the most_frequent data associated with that specific attribute
imputer = imputer.fit(X[:, 1:])
X[:, 1:] = imputer.transform(X[:, 1:])
```

```python
# Handling Categorial data
labelencoder_X = LabelEncoder()
X[:, 0] = labelencoder_X.fit_transform(X[:, 0])
onehotencoder = OneHotEncoder(drop='if_binary')
X = onehotencoder.fit_transform(X).toarray()

labelencoder_Y = LabelEncoder()
Y = labelencoder_Y.fit_transform(Y)
```
